```python
import pandas as pd

df_list = []
for year in range(2010, 2024):
    df = pd.read_excel("dataset1.xlsx", sheet_name=str(year))
    df_list.append(df)

data = pd.concat(df_list, ignore_index=True)
data
```

```python
data = data.drop(columns=["kabupaten", "tahun", "provinsi"])
data
```

```python
data = data.dropna()
data
```

```python
from datetime import datetime

is_datetime_row = data.map(lambda x: isinstance(x, datetime)).any(axis=1)
data = data[is_datetime_row == False]
data
```

```python
mask = data.apply(lambda row: '-' not in row.values, axis=1)
data = data[mask]
data = data.reset_index()
data = data.drop(columns=["index"])
```

```python
data = data.apply(pd.to_numeric, errors='coerce')
```

```python
data.info()
```

```python
data['angka_harapan_hidup'] = (data['angka_harapan_hidup_laki'] + data['angka_harapan_hidup_perempuan']) / 2
data['rerata_lama_sekolah'] = (data['rerata_lama_sekolah_laki'] + data['rerata_lama_sekolah_perempuan']) / 2
data
```

```python
data = data.drop(columns=['angka_harapan_hidup_laki', 'angka_harapan_hidup_perempuan', 'rerata_lama_sekolah_laki', 'rerata_lama_sekolah_perempuan'])
data
```
